    1. 演示操作过程中，自身分支在不断产生新的节点，类似于分支版本高于主干版本，导致无法进行同步（也就是版本回退）。
    2. 分支如果同步过最新的主干，将无法进行第二次同步工作。    
2. 对此，在解决过程中，出现了请求过多次，致使主干接收请求后出现文件丢失的情况，在一番处理后，主干回退到先前节点，总结下解决方法如下：
    1. 主干进行新的更新，产生新的主干节点供其他分支进行第二次同步。
    2. 分支进行请求回退，撤回到最开始的节点。

    - 将不需要刷新的属性赋值语句包在 if(!IsPostBack){}中
    - 如果session无法提取，清理浏览器缓存，重启项目
    - 导入包using System.Web.UI.HtmlControls;定义变量语句HtmlGenericControl a1 = new HtmlGenericControl("a");（其中a为超链接或锚标签）

    - 前端传递方式为href="xxx.aspx?name=001"；后台新建a标签传递参数方式为a1.Attributes.Add("href", string.Format("xxx.aspx?name={0}", u.name)); 接收传递到该页面的参数属性值的方式为：string name = Request.QueryString\["name"\];返回值类型为string

4. Java的list函数removeAll方法需要重载函数equals和hashcode方法。

```java
package exam;

import java.util.ArrayList;
import java.util.List;

public class User {

    private int id;
    private String name;

    public User() {}
    public User(int id, String name) {
        super();
        this.id = id;
        this.name = name;
    }
    public int getId() {
        return id;
    }
    public void setId(int id) {
        this.id = id;
    }
    public String getName() {
        return name;
    }
    public void setName(String name) {
        this.name = name;
    }

    @Override
    public String toString() {
        return this.id + " " + this.name;
    }
    
    @Override
    public int hashCode() {
        final int prime = 31;
        int result = 1;
        result = prime * result + id;
        result = prime * result + ((name == null) ? 0 : name.hashCode());
        return result;
    }
    @Override
    public boolean equals(Object obj) {
        if (this == obj)
            return true;
        if (obj == null)
            return false;
        if (getClass() != obj.getClass())
            return false;
        User other = (User) obj;
        if (id != other.id)
            return false;
        if (name == null) {
            if (other.name != null)
                return false;
        } else if (!name.equals(other.name))
            return false;
        return true;
    }
    public static void main(String[] args) {
        
        List<User> list = new ArrayList<User>();
        list.add(new User(1, "1"));
        list.add(new User(2, "2"));
        list.add(new User(3, "3"));
        list.add(new User(4, "4"));
        
        List<User> list1 = new ArrayList<User>();
        list1.add(new User(1, "1"));
        list1.add(new User(2, "2"));
        list1.add(new User(3, "3"));
        list1.add(new User(5, "5"));
        
        list.removeAll(list1);
        for (User u : list) {
            System.out.println(u.toString());
        }
        
        System.out.println();
        System.out.println(list.size());
    }
}
```

```SQL
CREATE sequence orders_seq --序列名
start with 1
increment by 1
minvalue 1
maxvalue 999999
no cache
no cycle

--调用
insert into orders values(next VALUE FOR ord_seq, 100, 123, GETDATE()); --GETDATE()得到当前时间

--查看所有已存在的sequence
select name from 数据库名.sys.sequences

--seclect当前的sequence值
select next value for orders_seq;
alter sequence orders_seq restart with next VALUE for orders_seq - 2; --因为每次调用next value for orders_seq都会使当前序列值+1，所以需要回退
```
